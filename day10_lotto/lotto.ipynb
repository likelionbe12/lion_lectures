```python
import requests

def getData(drwNo):
  url = f"https://www.dhlottery.co.kr/common.do?method=getLottoNumber&drwNo={drwNo}"
  response = requests.get(url)
  return response.json()
```

```python
drwNo=1141
dataList = []
while True:
  data = getData(drwNo)
  dataList.append(data)
  print(drwNo)
  drwNo+=1
```
